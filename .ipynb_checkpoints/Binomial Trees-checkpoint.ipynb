```python
import numpy as np
```

```python
def European_put(s, k, up, down, dt, T, r):
    # from ch 13.4
    # s: the current stock price
    # k: the strike price
    # up: the amount that the stock price will move up in percentage
    # down: the amount that the stock price will move down in percentage
    # dt: the duration of each step (in months)
    # T: the duration of all steps (in years)
    # r: the risk-free rate 

    u = 1+up*0.01
    d = 1-down*0.01
    dt = dt/12
    p = (np.exp(r*dt)-u)/(u-d)
    N = T/dt + 1

    stocks, options = [], []

    for n in N:
        stock, option = []
        for i in range(N-n):
            x = s*u**i*d**(N-i-1)
            if n == 0:
                y = k-x
            else:
                y = p*options[0]
```
